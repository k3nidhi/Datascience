```python
import pandas as pd

# Importing CSV file
df = pd.read_csv('/content/Spotify Most Streamed Songs.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot for categorical variables ('artist_count')
plt.figure(figsize=(10, 6))
sns.countplot(x='artist_count', data=df)
plt.title('Distribution of Artist Count')
plt.xlabel('Artist Count')
plt.ylabel('Frequency')
plt.show()

```

```python

# Set a colorful theme for categorical bar plots
sns.set_palette("Set2")  # You can change to other palettes like 'husl', 'coolwarm', 'Paired', etc.

# Bar plot for 'artist_count' (categorical variable)
plt.figure(figsize=(10, 6))
sns.countplot(x='artist_count', data=df, palette="Set2")  # Set2 palette for color
plt.title('Distribution of Artist Count', fontsize=15)
plt.xlabel('Artist Count', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
# List of continuous features to visualize
continuous_features = ['danceability_%', 'energy_%', 'valence_%',
                       'acousticness_%', 'instrumentalness_%',
                       'liveness_%', 'speechiness_%']

# Set a colorful palette (You can choose from various palettes: 'husl', 'Set2', 'coolwarm', etc.)
palette = sns.color_palette("husl", len(continuous_features))

# Plot histograms for each continuous feature with colors
for idx, feature in enumerate(continuous_features):
    plt.figure(figsize=(10, 6))
    sns.histplot(df[feature], bins=30, kde=True, color=palette[idx])
    plt.title(f'Distribution of {feature}', fontsize=15)
    plt.xlabel(feature, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.show()
```
